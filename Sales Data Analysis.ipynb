### question no 1- what was the best month for sales?how much was earned that month

```python
# but for sales we have to understand that we have  two coloums price and quantity we have to  make them one
# by multiplying them and make a new column sales
```

```python
df['sales']=df['Quantity Ordered'].astype('float')*df['Price Each'].astype('float')
```

```python
df
```

```python
dd=df.groupby('Ordermonth').sum()
dd.reset_index(inplace=True)
dd['sales']
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.barplot(x=dd['Ordermonth'],y =dd['sales'],data=dd)
```

### question2- what city has the highest no  of sales

```python
# for this question also we have to make a new column as cit
```

```python
df['city']=df['Purchase Address'].apply(lambda x:x.split(',')[1])
```

```python
df['state']= df['Purchase Address'].apply(lambda x:x.split(',')[2][0:3])
```

```python
df['citystate']= df['state']+df['city']
```

```python
df.drop(['city','state'],axis=1,inplace=True)

#dl['allstatecity']= dl.state + dl.city
```

```python
df
```

```python
dd=df.groupby('citystate')
dl=dd.sales.sum()
dl=dl.reset_index()
dl
```

```python
plt.bar(dl['citystate'],dl['sales'])
plt.xticks(dl['citystate'],rotation='vertical',size=8)
```

### question 3 - what time should we display advertisement to  maximize likelihood of customer buying product

```python
df['Order Date']=pd.to_datetime(df['Order Date'])
```

```python
df['Hour']=df['Order Date'].dt.hour
```

```python
dd=df.groupby('Hour').count()
dd=dd.reset_index()
dd
```

```python
sns.set(style='whitegrid')
sns.lineplot(x='Hour',y='sales',data=dd)
```
